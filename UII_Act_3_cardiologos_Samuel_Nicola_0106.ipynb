#### Redaccion del problema:Crear un programa que permita administrar una lista de médicos especialistas en cardiología.

#### menu
1.  Agregar
2.  Modificar
3.  Borrar
4.  Ver listas
5.  Salir



```python
Cardiologos = []

while True:
  print("1 Agregar elemento a la lista")
  print("2 Modificar elemento de la lista")
  print("3 Borrar elemento de la lista")
  print("4 Mostrar elementos de la lista ")
  print("5 Salir")

  Opcion = int(input("Inserte la opcion a realizar: "))

  if Opcion == 1:
    Nombre = input("Inserte el nombre del cardiólogo que desea agregar: ")
    Cardiologos.append(Nombre)

  elif Opcion == 2:
    Posicion = int(input("Inserte la posicion del cardiólogo que desea modificar: "))
    Nombre = input("Inserte el nuevo nombre del cardiólogo: ")
    Cardiologos[Posicion] = Nombre

  elif Opcion == 3:
    Posicion = int(input("Inserte la posicion del cardiólogo que desea borrar: "))
    Cardiologos.pop(Posicion)

  elif Opcion == 4:
    for X in Cardiologos:
      print(X)

  elif Opcion == 5:
    break

  else:
    print("Opcion no valida")

print("Programa realizado por Samuel Nicola NL = 40")
```
